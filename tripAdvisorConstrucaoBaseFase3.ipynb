```python
# Importando todas as bases sujas dos restaurantes:
df_restaurantes_sujo_parte1 = pd.read_csv('df_restaurantes_completo_sujo_parte1.csv')
df_restaurantes_sujo_parte2 = pd.read_csv('df_restaurantes_completo_sujo_parte2.csv')
df_restaurantes_sujo_parte3 = pd.read_csv('df_restaurantes_completo_sujo_parte3.csv')
df_restaurantes_sujo_parte4 = pd.read_csv('df_restaurantes_completo_sujo_parte4.csv')
df_restaurantes_sujo_parte5 = pd.read_csv('df_restaurantes_completo_sujo_parte5.csv')
df_restaurantes_sujo_parte6 = pd.read_csv('df_restaurantes_completo_sujo_parte6.csv')
df_restaurantes_sujo_parte7 = pd.read_csv('df_restaurantes_completo_sujo_parte7.csv')
df_restaurantes_sujo_parte8 = pd.read_csv('df_restaurantes_completo_sujo_parte8.csv')
df_restaurantes_sujo_parte9 = pd.read_csv('df_restaurantes_completo_sujo_parte9.csv')
df_restaurantes_sujo_parte10 = pd.read_csv('df_restaurantes_completo_sujo_parte10.csv')
df_restaurantes_sujo_parte11 = pd.read_csv('df_restaurantes_completo_sujo_parte11.csv')
df_restaurantes_sujo_parte12 = pd.read_csv('df_restaurantes_completo_sujo_parte12.csv')
df_restaurantes_sujo_parte13 = pd.read_csv('df_restaurantes_completo_sujo_parte13.csv')
df_restaurantes_sujo_parte14 = pd.read_csv('df_restaurantes_completo_sujo_parte14.csv')
df_restaurantes_sujo_parte15 = pd.read_csv('df_restaurantes_completo_sujo_parte15.csv')
df_restaurantes_sujo_parte16 = pd.read_csv('df_restaurantes_completo_sujo_parte16.csv')
df_restaurantes_sujo_parte17 = pd.read_csv('df_restaurantes_completo_sujo_parte17.csv')
df_restaurantes_sujo_parte18 = pd.read_csv('df_restaurantes_completo_sujo_parte18.csv')
df_restaurantes_sujo_parte19 = pd.read_csv('df_restaurantes_completo_sujo_parte19.csv')
df_restaurantes_sujo_parte20 = pd.read_csv('df_restaurantes_completo_sujo_parte20.csv')
df_restaurantes_sujo_parte21 = pd.read_csv('df_restaurantes_completo_sujo_parte21.csv')

# Criando uma lista de databases para concatená-las e salvá-las em um arquivo csv:
dfs = [df_restaurantes_sujo_parte1,
       df_restaurantes_sujo_parte2,
       df_restaurantes_sujo_parte3,
       df_restaurantes_sujo_parte4,
       df_restaurantes_sujo_parte5,
       df_restaurantes_sujo_parte6,
       df_restaurantes_sujo_parte7,
       df_restaurantes_sujo_parte8,
       df_restaurantes_sujo_parte9,
       df_restaurantes_sujo_parte10,
       df_restaurantes_sujo_parte11,
       df_restaurantes_sujo_parte12,
       df_restaurantes_sujo_parte13,
       df_restaurantes_sujo_parte14,
       df_restaurantes_sujo_parte15,
       df_restaurantes_sujo_parte16,
       df_restaurantes_sujo_parte17,
       df_restaurantes_sujo_parte18,
       df_restaurantes_sujo_parte19,
       df_restaurantes_sujo_parte20,
       df_restaurantes_sujo_parte21]

df_restaurantes_sujo = pd.concat(dfs)
df_restaurantes_sujo.to_csv('df_restaurantes_sujo.csv')

# Dropando linhas duplicadas:
df_restaurantes_sujo.drop_duplicates(inplace=True)

# Restartando index do dataframe:
df_restaurantes_sujo.reset_index(drop=True, inplace=True)
```

```python
# Usando regex para separar as informações das colunas 'notas' e 'detalhes':
# COLUNA NOTAS:
# Abrindo um dataframe vazio para incluir informações das colunas notas:
notas_df = pd.DataFrame()

for i in range(len(df_restaurantes_sujo)):
    
    # Variável notas é composta de um dicionário com notas médias para cada caracteristica do restaurante (comida, serviço, preço, ambiente):
    temp_notas_obs = df_restaurantes_sujo['notas'][i]
    # Extraindo nome da característica:
    nome_variavel = re.findall(r"'\w+':", temp_notas_obs)
    # Extraindo nota média da característica
    valor_variavel = re.findall(r"'\w+_\w+'", temp_notas_obs)
    
    if nome_variavel and valor_variavel is not None:
        
        # Criando um dicionário para inserir notas separadas em um dataframe que será concatenado ao dataframe dos restaurantes:
        nome_valor_variavel = {}
        for i in range(len(nome_variavel)):
            nome_valor_variavel[str(nome_variavel[i][1:-2])] = str(valor_variavel[i][-3:-1])   
        notas_df = notas_df.append(nome_valor_variavel, ignore_index=True)
        
    else:
        # Caso não exista informação na observação incluir uma linha de NaN
        index = pd.RangeIndex(1)
        linha_vazia = pd.DataFrame(np.nan, index=index, columns=notas_df.columns)
        notas_df = notas_df.append(linha_vazia, ignore_index=True)
        
# COLUNA DETALHES:
detalhes_df = pd.DataFrame()

for i in range(len(df_restaurantes_sujo)):
    
    # Variável notas é composta de um dicionário com valores para cada detalhe do restaurante (faixa de preço, cozinha, dietas especiais):
    temp_detalhes_obs = df_restaurantes_sujo['detalhes'][i]
    nome_variavel_detalhes = re.findall(r"'[a-zA-ZçÇãõ ]+':", temp_detalhes_obs)
    valor_variavel_detalhes = re.findall(r": '[a-zA-Z0-9 $çÇáãâéêíîóõôúûÁÃÂÉÊÍÎÓÕÔÚÛ,-\\]+'",temp_detalhes_obs)
    
    if nome_variavel_detalhes and valor_variavel_detalhes is not None:
        nome_valor_variavel = {}
        for i in range(len(nome_variavel_detalhes)):
            nome_valor_variavel[str(nome_variavel_detalhes[i][1:-2])] = str(valor_variavel_detalhes[i][3:-1])
        detalhes_df = detalhes_df.append(nome_valor_variavel, ignore_index=True)
        
    else:
        index = pd.RangeIndex(1)
        linha_vazia = pd.DataFrame(np.nan, index=index, columns=detalhes_df.columns)
        detalhes_df = detalhes_df.append(linha_vazia, ignore_index=True)
               
# Juntando as bases:
df_restaurantes_sujo = pd.concat([df_restaurantes_sujo, notas_df, detalhes_df], axis=1, join="inner")
```

```python
# COLUNA FAIXA DE PREÇOS:

# Criando duas colunas para incluir menor e maior preço praticado no restaurante:
df_restaurantes_sujo['preco_inicial'] = ''
df_restaurantes_sujo['preco_final'] = ''

for i in range(len(df_restaurantes_sujo)):
    
    # Preço inicial está na mesma "célula" da coluna faixa de preço, usando regex para separar o preço inicial e final:
    if str(df_restaurantes_sujo['FAIXA DE PREÇO'][i]) != str('nan'):
    
        temp_faixa_de_preco_obs = str(df_restaurantes_sujo['FAIXA DE PREÇO'][i])
        lista_precos = re.split(r"[a-zA-Z $\\-]+", temp_faixa_de_preco_obs)
        df_restaurantes_sujo['preco_inicial'][i] = lista_precos[1][1:]
        df_restaurantes_sujo['preco_final'][i] = lista_precos[2][1:]
                
    else:
        df_restaurantes_sujo['preco_inicial'][i] = 0
        df_restaurantes_sujo['preco_final'][i] = 0
```

```python
# Criando um dicionário para substituir caracteres das observações de cada variável:
subs = {" ": "_",
        "-": "_",
        "á": "a",
        "ã": "a",
        "â": "a",
        "é": "e",
        "ê": "e",
        "í": "i",
        "ó": "o",
        "õ": "o",
        "ú": "u",
        "ç": "c"}

# COLUNA COZINHAS:
# Inserindo uma vírgula antes dos caracteres da coluna para dividi-las em dummies posteriormente
df_restaurantes_sujo['COZINHAS'] = ', ' + df_restaurantes_sujo['COZINHAS']
# Transformando a variável 'COZINHAS' em uma variável dummy:
dummies_cozinhas = df_restaurantes_sujo['COZINHAS'].str.get_dummies(sep=",")
# Retirando caixa alta do nome das colunas:
dummies_cozinhas.columns = dummies_cozinhas.columns.str.lower()
# Transformando nomes das colunas em uma lista para incluir as iniciais "coz" em cada um deles:
nomes_cozinhas = dummies_cozinhas.columns.values.tolist()
nomes_cozinhas_1 = [str('coz') + re.sub(".", lambda m: subs.get(m.group(), m.group()), cozinha) for cozinha in nomes_cozinhas]
# Substituindo nomes antigos por novos:
dummies_cozinhas.columns = nomes_cozinhas_1

# COLUNA REFEIÇÕES:
df_restaurantes_sujo['Refeições'] = ', ' + df_restaurantes_sujo['Refeições']
dummies_refeicoes = df_restaurantes_sujo['Refeições'].str.get_dummies(sep=",")
dummies_refeicoes.columns = dummies_refeicoes.columns.str.lower()
nomes_refeicoes_lista = dummies_refeicoes.columns.values.tolist()
nomes_refeicoes = [str('ref') + re.sub(".", lambda m: subs.get(m.group(), m.group()), ref) for ref in nomes_refeicoes_lista]
dummies_refeicoes.columns = nomes_refeicoes

# COLUNA DIETAS ESPECIAIS:
df_restaurantes_sujo['Dietas especiais'] = ', ' + df_restaurantes_sujo['Dietas especiais']
dummies_dietas = df_restaurantes_sujo['Dietas especiais'].str.get_dummies(sep=",")
dummies_dietas.columns = dummies_dietas.columns.str.lower()
nomes_dietas_lista = dummies_dietas.columns.values.tolist()
nomes_dietas = [str('dietas') + re.sub(".", lambda m: subs.get(m.group(), m.group()), dieta) for dieta in nomes_dietas_lista]
dummies_dietas.columns = nomes_dietas

# COLUNA VANTAGENS:
df_restaurantes_sujo['VANTAGENS'] = ', ' + df_restaurantes_sujo['VANTAGENS']
dummies_vantagens = df_restaurantes_sujo['VANTAGENS'].str.get_dummies(sep=",")
dummies_vantagens.columns = dummies_vantagens.columns.str.lower()
nomes_vantagens_lista = dummies_vantagens.columns.values.tolist()
nomes_vantagens = [str('vantagem') + re.sub(".", lambda m: subs.get(m.group(), m.group()), vantagem) for vantagem in nomes_vantagens_lista]
dummies_vantagens.columns = nomes_vantagens


df_restaurantes_sujo = pd.concat([df_restaurantes_sujo, 
                                  dummies_cozinhas, 
                                  dummies_refeicoes,
                                  dummies_dietas,
                                  dummies_vantagens], axis=1)
```

```python
# COLUNA ENDEREÇO:
# Mantendo apenas os ceps na coluna endereço:
for i in range(len(df_restaurantes_sujo)):
    
    if str(df_restaurantes_sujo['endereco'][i]) != str('nan'):
        df_restaurantes_sujo['endereco'][i] = re.findall(r"[0-9]{5}-[0-9]{3}", df_restaurantes_sujo['endereco'][i])
    else:
        df_restaurantes_sujo['endereco'][i] = str('nan')
        
    if df_restaurantes_sujo['endereco'][i] != []:
        df_restaurantes_sujo['endereco'][i] = df_restaurantes_sujo['endereco'][i][0]
    else:
        df_restaurantes_sujo['endereco'][i] = str('nan')
        
# COLUNA PONTUACÃO:
# Retirando os pontos da coluna pontuação:
    if str(df_restaurantes_sujo['pontuacao'][i]) != str('nan'):
        df_restaurantes_sujo['pontuacao'][i] = re.findall(r"[0-9]+", df_restaurantes_sujo['pontuacao'][i])
        df_restaurantes_sujo['pontuacao'][i] = df_restaurantes_sujo['pontuacao'][i][0]+df_restaurantes_sujo['pontuacao'][i][1]
    else:
        df_restaurantes_sujo['pontuacao'][i] = str('0')
        
# COLUNA N_AVALIACOES:
# Retirando a palavra avaliações
    df_restaurantes_sujo['n_avaliacoes'][i] = re.findall(r"[0-9]+", df_restaurantes_sujo['n_avaliacoes'][i])[0]
```

```python
# Excluindo e renomeando colunas:
df_restaurantes_sujo.drop(columns=["Unnamed: 0", 
                                   "url", 
                                   "preco", 
                                   "notas", 
                                   "detalhes", 
                                   "FAIXA DE PREÇO",
                                   "COZINHAS", 
                                   "Refeições", 
                                   "Dietas especiais",
                                   "VANTAGENS"], axis=1, inplace = True)

df_restaurantes_sujo.rename(columns = {'endereco':'cep',
                                       'Comida': 'pontuacao_comida',
                                       'Serviço': 'pontuacao_servico',
                                       'Preço': 'pontuacao_preco',
                                       'Ambiente': 'pontuacao_ambiente'}, inplace=True)

# Substituindo observações NaN por 0 nas colunas de pontuação:
df_restaurantes_sujo[['pontuacao', 
                      'pontuacao_comida', 
                      'pontuacao_servico', 
                      'pontuacao_preco', 
                      'pontuacao_ambiente']].replace(0, np.nan, inplace=True)

# Transformando o tipo de variável de string para númerica:
df_restaurantes_sujo[['pontuacao', 
                      'n_avaliacoes', 
                      'pontuacao_comida', 
                      'pontuacao_servico', 
                      'pontuacao_preco', 
                      'pontuacao_ambiente', 
                      'preco_inicial', 
                      'preco_final']] = df_restaurantes_sujo[['pontuacao', 
                                                              'n_avaliacoes', 
                                                              'pontuacao_comida', 
                                                              'pontuacao_servico', 
                                                              'pontuacao_preco', 
                                                              'pontuacao_ambiente', 
                                                              'preco_inicial', 
                                                              'preco_final']].apply(pd.to_numeric)

#Dividindo todos os valores por 10, já que todos estão multiplicados por esse número:
df_restaurantes_sujo[['pontuacao', 
                      'pontuacao_comida', 
                      'pontuacao_servico', 
                      'pontuacao_preco', 
                      'pontuacao_ambiente']] = df_restaurantes_sujo[['pontuacao', 
                                                                     'pontuacao_comida', 
                                                                     'pontuacao_servico', 
                                                                     'pontuacao_preco', 
                                                                     'pontuacao_ambiente']]/10

df_restaurantes_sujo['preco_medio'] = (df_restaurantes_sujo['preco_final']+df_restaurantes_sujo['preco_inicial'])/2
```

```python
# Importando base de dados com os CEPs do Brasil (atualizada em 2018):
ceps = pd.read_csv('ceps.txt', sep="\t", header=None, names=["cep", "cidade_estado", "bairro", "rua", "obs"])
# Retirando hífen das observações da variável e substituindo todas as observações "nan" e "n" por 0:
df_restaurantes_sujo['cep'] = [df_restaurantes_sujo['cep'][i].replace("-", "") for i in range(len(df_restaurantes_sujo))]
df_restaurantes_sujo['cep'] = [df_restaurantes_sujo['cep'][i].replace("nan", "0") for i in range(len(df_restaurantes_sujo))]
df_restaurantes_sujo['cep'] = [df_restaurantes_sujo['cep'][i].replace("n", "0") for i in range(len(df_restaurantes_sujo))]

# Transformando os tipos das colunas em inteiros. Elas serão usadas para fazer o merge entre as bases:
df_restaurantes_sujo['cep'] = df_restaurantes_sujo['cep'].astype('int')
ceps['cep']=ceps['cep'].astype('int')

df_restaurantes_sujo = df_restaurantes_sujo.merge(ceps, how='inner', on='cep')

df_restaurantes_sujo.drop(columns=["cidade_estado",
                                   "rua",
                                   "obs"], axis=1, inplace = True)
```
